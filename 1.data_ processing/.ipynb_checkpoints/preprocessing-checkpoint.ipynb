## Data preprocessing and cutting words

```python
import re
import pandas as pd
import numpy as np
import os
from os import listdir
import jieba
import pickle
from collections import Counter
```

```python
os.chdir('../')
os.getcwd()
```

## 取資料夾內容

```python
news_path = '/project/jt071-group23/news'
files = listdir(news_path)
```

```python
if '.ipynb_checkpoints' in files:
    files.remove('.ipynb_checkpoints')
files
```

```python
key = []
value = []

for i,foldName in enumerate(files):
    if foldName != '.ipynb_checkpoints':
        key.append(foldName)
        value.append(str(i))
        print(i,foldName)
    
# dictionary for label mapping
label_mapping = dict(zip(key,value))
label_value_key = dict(zip(value, key))
```

```python
label_value_key
```

#### 合成csv檔案路徑

```python
file_path = []
for f in files:
    if f !='.ipynb_checkpoints':
        path = '/project/jt071-group23/news/'+ f 
        file_path.append(path)
```

```python
# data = pd.read_csv(file_path[0])
# test = data['content'][2]
# test
```

#### loading data and label

```python
data_content = []
data_label = []

for path in file_path:
    data = pd.read_csv(path)
    
    # 刪除類別中重複的新聞
#     data_drop = data.drop_duplicates('content', keep='first')
#     fileName = path.split('/')[-1]
#     print(fileName, '|', data_drop.shape[0])
    
    # 將每個類別中的content加到all_content
    for content in data['content']:
        if content is not np.nan:
            data_content.append(content)
            
            # 取出每個content的類別
            fileName = path.split('/')[-1]
            label = label_mapping.get(fileName)
            data_label.append(label)
        
        
print('data_content',len(data_content))
```

```python
df_all_data = pd.DataFrame({'label':data_label, 'content':data_content})
df_all_data.head()
print(df_all_data.shape)
```

```python
df_all_data['label_name'] = df_all_data['label'].map(label_value_key)
df_all_data.head()
```

```python
df_all_data['content'].isnull().value_counts()
```

#### overview data

```python
df_data_calculate = df_all_data.groupby(['content','label_name']).first().unstack()
```

```python
df_data_calculate.shape
```

```python
df_data_calculate['label_times'] = df_data_calculate.count(axis=1, level='label_name').sum(axis=1)
print(df_data_calculate.shape)
df_data_calculate.head()
```

```python
for i in range(1,15):
    number = df_data_calculate[df_data_calculate['label_times']==i].shape[0]
    print('文章分類為',i,'類有', number, '篇')
```

```python
# df_all_data.head()
```

## 刪除類別中重複的文章(81804)

```python
df = df_all_data.copy()
df = df.drop_duplicates('content', keep='first')
print(df.shape)
```

#### jieba cut words

```python
os.getcwd()
```

```python
jieba.set_dictionary('/project/jt071-group23/jt071078/jieba/dict.txt.big')
stop_words = open('/project/jt071-group23/jt071078/jieba/stop_words.txt').read().splitlines()
```

```python
Counter(pd.isnull(df['content']))
```

```python
df.drop(df[pd.isnull(df['content'])].index, inplace=True)
df.shape
```

```python
content_list = df['content'].tolist()

sentences = []

for i, text in enumerate(content_list):
    line = []

    for w in jieba.cut(text, cut_all=False):
        
        ## remove stopwords and digits
        ## can define your own rules
        if w not in stop_words and not bool(re.match('[0-9]+', w)):
            line.append(w)

    sentences.append(line)

    if i%10000==0:
        print(i, '/', len(content_list))
```

```python
df['sentences'] = sentences
```

```python
df['sentences_1000'] = df['sentences'].apply(lambda x:x[0:1000] if len(x)>1000 else x)
```

```python
df['sentences_1000_str'] = df['sentences_1000'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df.shape
```

```python
with open('/project/jt071-group23/jt071078/data/cut_words(81804)', 'wb') as file:
    pickle.dump(sentences, file)
```

```python
df.to_csv('/project/jt071-group23/jt071078/for_CNN_data(81804).csv', index=False)
```

## 取出分類為2的文章（11097）
1. 找出文章為哪兩個分類
2. 計算哪兩個分類的文章最多，進行重新分類

```python
df = df_data_calculate[df_data_calculate['label_times']==2].copy()
df.shape
```

```python
df.head()
```

```python
df = df.drop('label_times',axis=1)
```

```python
df = df.reset_index()
```

```python
df.head()
```

```python
# 增加label pair column
label_number = df['label'].values

label_number_list =[]

for i in range(label_number.shape[0]):
    number_array = label_number[i]
    number_pair = number_array[~pd.isnull(number_array)]
    label_number_list.append(tuple(number_pair))
    
df['label_pair'] = label_number_list
```

```python
# 計算哪兩種分類的文章數
# (4,9)為(美食,旅遊)有4306篇
# (10,3)為(理財,財經)有1597篇
# (6,0)為(3C,蘋果專區)有627篇
# (13,3)為(區塊鏈,財經)有626篇
# (9,7)為(美食,醫療)有581篇
```
