```python
%matplotlib inline

import re
import pandas as pd
import os
import seaborn as sns
import jieba
import jieba.posseg as pseg
```

```python
csv_path = '/home/jovyan/AIA_project/spider/news'
csv_list = os.listdir(csv_path)
df = pd.DataFrame()
for f in csv_list:
    if f !='.ipynb_checkpoints':
        class_no = f.split('_')[0]
        csv_data = pd.read_csv(csv_path+'/'+f)[['url','eigenWords','topKeyWords','content','title']]
        csv_data['class_no'] = int(class_no)
        df = pd.concat([df, csv_data], axis=0, ignore_index=True)
```

```python
df.shape
```

```python
df['content'] = df['content'].str.replace('https?:\/\/\S*', '')
```

```python
jieba.set_dictionary('jieba/dict.txt.big')
stop_words = open('jieba/stop_words.txt').read().splitlines()
```

```python
df.drop(df[pd.isnull(df['content'])].index, inplace=True)
```

```python
content_list = df['content'].tolist()

sentences = []

for i, text in enumerate(content_list):
    line = []

    for w in jieba.cut(text, cut_all=False):
        
        ## remove stopwords and digits
        ## can define your own rules
        if w not in stop_words and not bool(re.match('[0-9]+', w)):
            line.append(w)

    sentences.append(line)

    if i%10000==0:
        print(i, '/', len(content_list))
```

```python
df['sentences'] = sentences
```

```python
df['sentences_1000'] = df['sentences'].apply(lambda x:x[0:1000] if len(x)>1000 else x)
```

```python
df['sentences_1000_str'] = df['sentences_1000'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop(df[pd.isnull(df['title'])].index, inplace=True)
```

```python
title_list = df['title'].tolist()

t_sentences = []

for i, text in enumerate(title_list):
    line = []

    for w in jieba.cut(text, cut_all=False):
        
        ## remove stopwords and digits
        ## can define your own rules
        if w not in stop_words and not bool(re.match('[0-9]+', w)):
            line.append(w)

    t_sentences.append(line)

    if i%10000==0:
        print(i, '/', len(content_list))
```

```python
df['t_sentences'] = t_sentences
```

```python
df['title_str'] = df['t_sentences'].apply(lambda x:" ".join(x))
```

```python
df.to_csv('data/data_all_t.csv', index=False)
```
